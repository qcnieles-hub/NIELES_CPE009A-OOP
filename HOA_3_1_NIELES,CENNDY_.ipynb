The list numberlist has five elements: [5, 4, 2, 1, 3], so len(numberlist) returns 5. Using positive indexing, numberlist[0] gives the first element 5, numberlist[1] gives 4, numberlist[2] gives 2, numberlist[3] gives 1, and numberlist[4] gives 3. Attempting to access numberlist[5] results in an IndexError, because index 5 is out of range. With negative indexing, numberlist[-1] gives the last element 3, numberlist[-2] gives 1, numberlist[-3] gives 2, numberlist[-4] gives 4, and numberlist[-5] gives the first element 5. Trying to access numberlist[-6] also results in an IndexError, because -6 is outside the valid range. In summary, positive indices go from 0 to len(list)-1, negative indices go from -1 to -len(list), and accessing indices beyond these limits causes errors.

Distance Formula
1. Make a program that would calculate the distance between two points given a list of coordinates. Use the distance
formula.
coorindates_list = [(1,1), (2,3)]


```python
import math

coordinates_list = [(1, 1), (2, 3)]
(x1, y1) = coordinates_list[0]
(x2, y2) = coordinates_list[1]
distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
print("Distance between points:", distance)
```

Phonebook

3. Create a simple phonebook program that can read from a list of dictionaries. The program should be able to display a
person’s name, contact, and address based from a user input which is the id of the record.

```python
phonebook = [
    {'id': 1, 'name': 'Nieles Cenndy', 'contact': '09060611233', 'address': '123 Main St'},
    {'id': 2, 'name': 'Waffle Fustino', 'contact': '09123456789', 'address': '456 Oak Ave'},
    {'id': 3, 'name': 'Magic Fustino', 'contact': '09234567890', 'address': '789 Pine Rd'}
]

try:
    search_id = int(input("Enter ID to search: "))
except ValueError:
    print("Invalid input. Please enter a number for the ID.")
    search_id = None


found = False

if search_id is not None:
    for person in phonebook:
        if person['id'] == search_id:
            print("Name:", person['name'])
            print("Contact:", person['contact'])
            print("Address:", person['address'])
            found = True
            break

    if not found:
        print("No record found with ID:", search_id)
```
