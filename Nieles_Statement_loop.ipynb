LOOP STATEMENT

# Example 1 (The Break Statement)

```python
week=["Sunday","Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"]
for x in week:
    print(x)
```

# Example 2 (The Break Statement)

```python

week=["Sunday","Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"]
for x in week:
    print(x)
    if x=="Thursday":
        break
```

# Example 3 (The Break Statement)

```python
week = ["Sunday","Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"]

for x in week:
    if x == "Thursday":
        print(x)
        break

```

Challenge for Break Statement in For Loop Statement

```python
week = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
for x in week:
  if x == "Wednesday":
    break
  print(x)
```

The range () Function

```python
for x in range(6):
  print(x)
for x in range(2,6):
  print(x)
```

```python
for x in range(2,2,6):
    print(x)

```

Nested Loops

```python
adjectives = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

for x in adjectives:
    for y in fruits:
        print(x, y)

```

 # Example 5 While loop Statement

```python
i = 1
while i < 6:
  print(i)
  i += 1 #i = i+1 Assigment operator
```

```python
i = 1
while i < 6:
    print(i)
    if i == 3:
        break
    i += 1

```

The Continue Statement

```python
i = 0
while i < 6:
    i += 1
    if i == 3:
        continue
    print(i)

```

```python
i = 0
while i < 6:
    i += 1
    if i == 3:
      print(i)
      continue




```

The else Statement

```python
i = 1
while i < 6:
  print(i)

  i += 1
else:
  print("i is no longer less than 6")
```
