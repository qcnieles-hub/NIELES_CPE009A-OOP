Elif Statements

```python
a = 12
b = 12
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")

```

else statement

```python
a = 200
b = 33
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
else:
  print("a is greater than b")
```

```python
a = 200
b = 33
c = 500
if a > b and c > a:
  print("Both conditions are True")
```

```python
x = 41

if x > 10:
    print("Above ten,")
    if x > 20:
        print("and also above 20!")
        if x > 30:
            print("and also above 30")
        else:
            print("below 30")
    else:
        print("below 20.")
else:
    print("below 10")

```
