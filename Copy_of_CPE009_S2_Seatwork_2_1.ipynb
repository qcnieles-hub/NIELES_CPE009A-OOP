```python
# We want to create a python code for a grading scheme

# Create a function to accept a number of activities
# Compute for the score vs total
# Return the average grade of the student in %
def termCrit(actNumber):
  rawScore = []
  totalScore = []

  for i in range(0, actNumber):
    inputScore = float(input("Score: "))
    inputTotal = float(input("Total: "))

    if inputScore > inputTotal:
      print("\nInvalid Scores.\n")
      inputScore, inputTotal = 0,0
      break
    else:
      rawScore.append(inputScore)
      totalScore.append(inputTotal)

  grade = sum(rawScore) / sum(totalScore)
  grade = grade * 100
  return grade
```

```python
# Create a function for the class standing per term
def termCS():
  print("Hands-on Activities: ")
  termHOA = termCrit(int(input("How many hands-on activities this term?")))

  print("Quizzes:")
  termQuiz = termCrit(int(input("How many quizzes this term?")))

  print("Assignment: ")
  termAss = termCrit(int(input("How many assignments?")))

  grade = (0.5 * termHOA) + (0.3 * termQuiz) + (0.2 * termAss)
  return grade
```

```python
# Function to compute average percentage for a criterion
def termCrit(count):
    if count == 0:
        return 0

    totalPercent = 0

    for i in range(count):
        score = float(input(f" Score #{i+1}: "))
        total = float(input(f" Total #{i+1}: "))
        percent = (score / total) * 100
        totalPercent += percent

    return totalPercent / count


# Function for class standing per term
def termCS():
    print("Hands-on Activities:")
    termHOA = termCrit(int(input("How many hands-on activities this term? ")))

    print("Quizzes:")
    termQuiz = termCrit(int(input("How many quizzes this term? ")))

    print("Assignments:")
    termAss = termCrit(int(input("How many assignments? ")))

    grade = (0.5 * termHOA) + (0.3 * termQuiz) + (0.2 * termAss)
    return grade


# Function to compute the term grade
def termGrade():
    print("Computing Term Grade")

    examScore = float(input("Exam Score: "))
    examTotal = float(input("Exam Total: "))
    examGrade = (examScore / examTotal) * 100

    CSgrade = termCS()

    termGrade = (0.5 * examGrade) + (0.5 * CSgrade)
    return termGrade


# Function to compute semester grade
def semesterGrade():
    print("\nPrelim Grade")
    prelim = termGrade()

    print("\nMidterm Grade")
    midterm = termGrade()

    midtermFinal = (1/3) * prelim + (2/3) * midterm

    print("\nFinal Grade")
    final = termGrade()

    semesterFinal = (1/3) * midtermFinal + (2/3) * final
    return semesterFinal


# Function to convert numeric grade to grade equivalent
def convertGrade(finalGrade):
    if finalGrade >= 94:
        return 1.0
    elif finalGrade >= 88:
        return 1.25
    elif finalGrade >= 82:
        return 1.5
    elif finalGrade >= 76:
        return 1.75
    elif finalGrade >= 70:
        return 2.0
    elif finalGrade >= 64:
        return 2.5
    elif finalGrade >= 58:
        return 3.0
    else:
        return 5.0


# Driver Code
finalGrade = semesterGrade()
print("\nFinal Semester Grade:", round(finalGrade, 2))
print("Grade Equivalent:", convertGrade(finalGrade))
```
