```python
def compute_cs(hoa, quiz, assignment):
    """Compute Class Standing"""
    return hoa * 0.5 + quiz * 0.3 + assignment * 0.2


def tips_grade(final_grade):
    """Convert Final Grade to T.I.P.S Numerical Grade"""
    if final_grade >= 96:
        return 1.00
    elif final_grade >= 94:
        return 1.25
    elif final_grade >= 92:
        return 1.50
    elif final_grade >= 89:
        return 1.75
    elif final_grade >= 87:
        return 2.00
    elif final_grade >= 84:
        return 2.25
    elif final_grade >= 82:
        return 2.50
    elif final_grade >= 79:
        return 2.75
    elif final_grade >= 75:
        return 3.00
    else:
        return 5.00


def compute_student():
    """Compute grades for one student"""

    # INPUT STUDENT NAME (ONCE ONLY)

    name = input("\nEnter Student Name: ")

    # PRELIM
    print("\n--- PRELIM ---")
    print("Student Name:", name)

    phoa = float(input("Prelim HOA: "))
    pquiz = float(input("Prelim Quiz: "))
    pas = float(input("Prelim Assignment: "))
    pcs = compute_cs(phoa, pquiz, pas)

    pexam = float(input("Prelim Exam: "))
    prelim_grade = (pexam * 0.5) + (pcs * 0.5)

    # MIDTERM
    print("\n--- MIDTERM ---")
    print("Student Name:", name)

    mhoa = float(input("Midterm HOA: "))
    mquiz = float(input("Midterm Quiz: "))
    mas = float(input("Midterm Assignment: "))
    mcs = compute_cs(mhoa, mquiz, mas)

    mexam = float(input("Midterm Exam: "))
    midterm_exam_grade = (mexam * 0.5) + (mcs * 0.5)
    midterm_grade = (prelim_grade / 3) + (2 * midterm_exam_grade / 3)

    # FINAL
    print("\n--- FINAL ---")
    print("Student Name:", name)

    fhoa = float(input("Final HOA: "))
    fquiz = float(input("Final Quiz: "))
    fas = float(input("Final Assignment: "))
    fcs = compute_cs(fhoa, fquiz, fas)

    fexam = float(input("Final Exam: "))
    final_exam_grade = (fexam * 0.5) + (fcs * 0.5)
    final_grade = (midterm_grade / 3) + (2 * final_exam_grade / 3)

    tips = tips_grade(final_grade)

    # DISPLAY RESULTS
    print("\n--- STUDENT GRADE REPORT ---")
    print("Name:", name)
    print("Prelim Grade:", f"{prelim_grade:.2f}")
    print("Midterm Grade:", f"{midterm_grade:.2f}")
    print("Final Grade:", f"{final_grade:.2f}")
    print("T.I.P.S Numerical Grade:", tips)


# TEST 3 STUDENTS
for i in range(3):
    compute_student()
```
